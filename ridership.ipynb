```python
# import pandas and numpy module  
import pandas as pd  
import numpy as np


# making dataframe  
ridership = pd.read_csv("Ridership.csv")  
   
# output the dataframe 
ridership.head()
```

```python
new_names = (
    ridership.columns
    .str.replace(' #', '_num')
    .str.replace('-', '_') # this is for one problem making column `'community - board'`
    .str.replace(' ', '_')
    .str.lower()
)

if new_names.is_unique == 1:
    ridership.columns = new_names # updates my column names
ridership.columns

#replacing ridership 0 values with nan
ridership = ridership.replace("0", np.nan)
ridership = ridership.replace("NaN", np.nan)

ridership
```

```python
for col in ridership.columns:
    if ridership[col].dtype == 'object':  # Check if the column is of type object (string)
        ridership[col] = ridership[col].str.replace(',', '')

for col in ridership.columns:
    if ridership[col].name != "day_type":
        if ridership[col].name != "route":
            ridership[col] = ridership[col].astype(float)

ridership.info()
```
