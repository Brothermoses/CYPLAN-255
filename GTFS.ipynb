```python
#GTFS Data

# import gtfs_kit as gk
# feed = gk.read_feed('dec2024', dist_units='mi')

import pandas as pd

shapes_df = pd.read_csv('/workspaces/CYPLAN-255/GTFS/GTFS/shapes.txt')

#CHATGPT IS ALL USED BELOW
import geopandas as gpd
from shapely.geometry import Point, LineString
# Group shapes by shape_id and create LineStrings
lines = shapes_df.groupby('shape_id').apply(lambda x: LineString(x[['shape_pt_lon', 'shape_pt_lat']].values)).reset_index()
lines.columns = ['shape_id', 'geometry']
# Create a GeoDataFrame
geo_df = gpd.GeoDataFrame(lines, crs="EPSG:4326")
```

```python
geo_df.plot()
```
